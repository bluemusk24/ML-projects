#Seaborn is a python library used for statistical plot. It's built on Matplotlib and works with pandas dataframe.

# Distribution Plots: for uni-varaint(one variable) data

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#seaborn has built-in dataset that can be opened. example is tips.

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#use distplot() for a numerical column of the dataset tips. gives a histogram and KDE( kernel density estimation)

sns.distplot(tips['total_bill'])
```

```python
#remove kde from distplot and set bins(horizontal). Bins depend on dataset you working on

sns.distplot(tips['total_bill'],kde=False,bins=40)
```

# Joint Plots: works with distplot for bi-varaint data

```python
tips.head()
```

```python
#joint plots of 2 numerical variables from tips.head() gives a scattered plot.this is the default.

sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#same as above plot
sns.jointplot(y=tips['tip'],x=tips['total_bill'])
```

```python
#joint plot allows to change plots with kind parameter: hex

sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# reg for regression plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#kde kind for joint plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

# Pair plot: plot all numerical columns in a dataset.

```python
tips.head()
```

```python
# a histogram displays when a column is plotted against itself. scatter plot displays when different columns plot

sns.pairplot(tips)
```

```python
#'hue' parameter indicates a column where the categories are different. Males in sex column was different from females.

sns.pairplot(tips,hue='sex')
```

```python
#palette parameter makes the hue in the pairplot transparent

sns.pairplot(tips,hue='sex',palette='coolwarm')
```

# Rugplot: works like distplot for a uni-variant column. it draws a dashline for the chosen column.

```python
tips.head()
```

```python
sns.rugplot(tips['total_bill'])
```

```python
#difference between rugplot and distplot is all the marks are stashed together on a rugplot while distplot gives histogram

sns.rugplot(tips['total_bill']), sns.distplot(tips['total_bill'])
```

```python
#kde is the sum of normal distribution around the rugplot.

sns.kdeplot(tips['total_bill'])
```

# Categorical Plots: deal in non-numerical columns in relation to numerical columns.

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# barplot for categorical column in relation to a numerical column. it indicates average/mean for a category

sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
# estimator parameter to find standard variation,variance,sum or any arithmetric in a barplot

import numpy as np

sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#estimator to find the sum of each category in categorical column
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.sum)
```

#countplot: similar to barplot except the estimator is counting number of occurences.

```python
sns.countplot(x='sex',data=tips)
```

#boxplot: to show distribution based of categorical data

```python
# the interpretes as quartiles. the dots above are outliers
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# to add a hue to boxplot to differentiats smokers from non-smokers
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

#violin plot: works like boxplot to show distribution based off a category in a dataset

```python
#violin plot give more info than boxplot but difficult to read. boxplot is easier to read.
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#split=True joins the distribution in violin plot together and hue gives another data in the same plot.
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

#stripplot: just a scattered plot of a categorical and numerical columns

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
# similar to stripplot above without jitter
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

#swarm plot: similar to strip plot but the points don't overlap. strip plot combibed with violin plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

#Always use either boxplot, countplot or barplot to interprete data.

```python
#factorplot: most general of all categorical plots. it has kind parameter that takes anyother plot.
```

```python
sns.factorplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='strip')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='swarm')
```

# Matrix Plots: involves creating heatmap

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
# to get matrix plots, you need heatmap to have same or another variable name for both rows(index) and columns
tips.corr()
```

```python
#heatmap colors the values base off a gradient scale
sns.heatmap(tips.corr())
```

```python
# annotate fixes the correlated values of the dataset into heatmap.
sns.heatmap(tips.corr(),annot=True)
```

```python
# to change color of the heatmap
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
```

```python
flights
```

```python
# make flights rows to have an index name with matrix plot pivot table.
fc = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fc
```

```python
sns.heatmap(fc)
```

```python
# to add linewidth,linecolor and cmap to the pivot table
sns.heatmap(fc,linewidth=3,linecolor='white',cmap='magma')
```

```python
flights.index.names = ['numbers']   # not necessariy to seaborn, just to name an index.
```

```python
flights
```

```python
sns.heatmap(fc,linewidth=3,linecolor='black',cmap='coolwarm')
```

```python
#cluster map clusters rows and columns together randomly and how they match. It clusters base on hierachy
sns.clustermap(fc,cmap='coolwarm')
```

```python
# to set a standard scale for the clustering
sns.clustermap(fc,cmap='coolwarm',standard_scale=1)
```

# Regression Plots: applicable in machine learning. LM plots instead

```python
# LM plots allows display of linear models with seaborn.it uses some parameter from matplotlib too.
```

```python
tips.head()
```

```python
#to get data with lm plot
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# assign different sex as hue and markers to be specific for each sex.
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
# to access matplotlib on the markers on seaborn and make the hue marker bigger:
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
#separate the sexes male lm from female lm: gridplots of col and row
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
# to get the lm plot tip and total_bill for each sex in a particular day
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='day')
```

```python
#set the size and aspect for the above. aspect is the ratio btwn height and width of each grid
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='day',row='smoker',aspect=0.5,height=5)
```

# Grids: used to automate subplots based off data features.

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
```

```python
# iris dataset is a data of flowers measurement. more of this in machine learning.
iris.head()
```

```python
iris['species'].count(), iris['species'].unique(), iris['species'].nunique()
```

```python
#call automated joint data
sns.pairplot(iris,hue='species')
```

```python
sns.pairplot(iris)
```

```python
# to call PairGrid, it gives an empty Grid. works like pairplot but you control/customise the plot by yourself

g = sns.PairGrid(iris)
```

```python
#use map alongside different kinds of grids on PairGrid. histplot works like distplot

import matplotlib.pyplot as plt

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# FacetGrid: works like subplots Matplotlib

```python
tips.head()
```

```python
#works with map like too
g = sns.FacetGrid(tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
# to use scatter plot with FacetGrid. this involves two numerical columns
g.map(plt.scatter,'total_bill','tip')
```

# Seaborn Exercise and Solutions

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# call titanic data from seaborn and set style
titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')
```

```python
titanic.head()
```

```python
# to create a joint plot of titanic with age on y-axis and fare on x-axis
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
# to create a distplot for the the fare column
sns.distplot(titanic['fare'],kde=False,color='red',bins=30)
```

```python
# to create a box plot of titanic with age on y-axis and fare on x-axis
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
# to create a swarm plot of titanic with age on y-axis and fare on x-axis
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

```python
# to create a cout(bar) plot of titanic with sex as x-axis
sns.countplot(x='sex',data=titanic)
```

```python
# to create a matrix correlation plot
titan = titanic.corr()
```

```python
titan
```

```python
#to create heatmap for the matrix corr titan with annot
sns.heatmap(titan,cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
sns.heatmap(titan,annot=True,cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
# facetgrid with map function. I used hisplot instead of distplot
ship = sns.FacetGrid(titanic,col='sex')
ship.map(sns.histplot,'age')
```

```python
ship = sns.FacetGrid(titanic,col='sex')
ship.map(plt.hist,'age')
```
