```python
#all_banks data was downloaded from kaggle and opened as a dataframe with pd.read_pickle() 
df = pd.read_pickle("C:\\Users\\emman\\all_banks")
```

```python
df     #stocks of: BAC(Bank of America), C(Citi Group),GS(Goldman Sachs),JPM(JP Morgan) MS(Morgan Stanley), WFC(Wells Fargo)
```

```python
#create a list of Bank Ticker and call it tickers
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
#concatenate banks together as bank_stocks along the column axis and set key to tickers.
bank_stocks = pd.concat([df],axis=1)
```

```python
#what is the max close price of each bank's stock throughout time period?
bank_stocks['BAC']['Close'].max(),bank_stocks['C']['Close'].max(),bank_stocks['GS']['Close'].max(),bank_stocks['JPM']['Close'].max(),bank_stocks['MS']['Close'].max(),bank_stocks['WFC']['Close'].max()
```

```python
# ALSO finding max close price collectively,
for ticks in tickers:
    print(ticks,bank_stocks[ticks]['Close'].max())
```

```python
#using the xs() pandas method to get the max close price:

bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
#create a new datafram returns to contains returns for each bank stock. Use the return formula below
returns = pd.DataFrame()
```

```python
#use pandas pct_change() method to on the close column create column representing this return value. 
#create a for loop that goes for each Bank Stock Ticker. create the returns column and set the columns in return dataframe.

for tick in tickers:
    returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()     #first NaN values tells you can't get % value on the first day of stocks.
```

```python
#this can also be used to create returns column.
for tick in tickers:
    returns.append(bank_stocks[tick]['Close'].pct_change())
```

# MOVING AVERAGE

```python
#Plot the rolling 30 day average against the Close Price for Bank of America's Stock for year 2008
plt.figure(figsize=(10,5))
bank_stocks['BAC']['Close'].loc['2008'].rolling(window=30).mean().plot(label='30 day Moving Average')
bank_stocks['BAC']['Close'].loc['2008'].plot(label='BAC Close')
plt.legend(loc=2,bbox_to_anchor=(1.05,1))
```

```python
#create a heatmap of the correlation between the stocks Close price
bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr()
```

```python
#matrix form with corr() makes the index variables.
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```

```python
#cluster the above correlations together.
sns.clustermap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```

```python
#create a candle plot for Bank Of America stock from Jan 2015-Jan 2016
bank_stocks['BAC'].loc['2015'].iplot(kind='candle')
```

```python
#use technical analysis plot .ta_plot(study='sma') to create a Simple Moving Average plot of Morgan Stanley for year 2015.
bank_stocks['MS']['Close'].loc['2015'].ta_plot(study='sma',periods=[13,21,55])
```

```python
#use technical analysis plot .ta_plot(study='boll') to create a Bollinger Band plot for Bank of America for year 2015.
bank_stocks['BAC']['Close'].loc['2015'].ta_plot(study='boll')
```
